# Анализ изменения стоимости медицинского обслуживания

**Цель**

* Требуется провести анализ изменения стоимости лечения в 2022 году относительно 2021 года в медицинской клинике и определить ключевые драйверы изменения стоимости.

**Задание**
* Проверить следующие гипотезы:
1.	Средняя стоимость обслуживания пациентов изменилась из-за изменения цен на услуги.
2.	Средняя стоимость обслуживания изменилась из-за изменения интенсивности лечения пациентов.
3.	Средняя стоимость обслуживания пациентов изменилась из-за изменения поло-возрастной структуры обслуживаемых пациентов
4.	Средняя стоимость обслуживания пациентов изменилась из-за изменения структуры оказываемых услуг (раньше были дорогие услуги, теперь дешевые или наоборот)
5.	Средняя стоимость обслуживания изменилась из-за изменения среднего количества оказанных услуг на один кейс
6.	Другие гипотезы на усмотрение кандидата



**Описание данных**
* record_id- уникальный идентификатор строки данных
* service_date - дата оказания медицинской услуги
* service_name - наименование медицинской услуги
* service_number - количество услуг
* service_amount - сумма выплат (стоимость оказанных услуг в рублях)
* insured - уникальный идентификатор пациента
* sex_id - пол пациента
* age_for_service_date - возраст пациента

Все данные взяты из одной клиники



* **Медицинский кейс** - последовательность медицинских услуг, полученных одним застрахованном в одной клинике с интервалом между любыми соседними по дате услугами не более 14 дней. Т.е. между любыми 2-мя услугами в кейсе не должно быть разрыва более 14 дней.
* **Визит** - уникальное сочетание даты оказания услуги, застрахованного. Т.е. один визит - это один день, в который застрахованный побывал в клинике. Если был в одной клинике 2 раза, но в разные дни, то это два визита. Если получал услуги (в любом количестве) в одной клинике в один день, то это один визит.
* **Обратившийся в месяц** - уникальное сочетание месяца оказания услуги и идентификатора пациента

### Введение показателя расчёта стоимости лечения

* Создадим сводную таблицу для расчётов изменения стоимости лечения. 
* Таблица будет содержать информацию по медицинским кейсам (суммарные значения стоимости и количества услуг за  14 дней). 

```python
# Создадим таблицу с окном 14 дней

medical_cases = (clinic
                 .set_index('service_date')
                 .groupby(['insured', 'sex_id', 'age_for_service_date', 'year'])
                 .resample('14D')[['service_number', 'service_amount']]
                 .sum()
                 .reset_index()
                 .fillna(0)
                )



# Удалим пустые строки
medical_cases = medical_cases.query('service_number>0 and service_amount>0')

# Добавим колонку medical_case для нумерации медицинского кейса
medical_cases['medical_case'] = medical_cases.groupby('insured').cumcount() + 1

# Добавим колонку средней стоимости одной услуги
medical_cases['service_amount_per_one'] = medical_cases.service_amount/medical_cases.service_number



# Изменим формат
medical_cases = medical_cases.astype({'service_number': 'int'})


# Выведем таблицу
medical_cases.head()
```

* В то время, как максимальная стоимость выросло с 69 тыс до 105 тыс рублей за один медицинский кейс.
* Проверим гипотезу о равности двух выборок.

```python
# Проверим на нормальность распределения данных стоимости обслуживания пациентов с помощью теста Колмогорова-Смирнова
stats.kstest(medical_cases['service_amount'],'norm')
```

* Данные распределены ненормально, значит будем применять непараметрический тест Манна-Уитни для попарного сравнения двух независимых выборок.

```python
# Проверим одностороннюю гипотезу, что стоимость обслуживания пациентов меньше в 2021 году, чем в 2022 г.   
alpha = .05 # уровень значимости
results = stats.mannwhitneyu(x=medical_cases.query('year==2021')['service_amount'],
                   y=medical_cases.query('year==2022')['service_amount'],
                  alternative='less') # результаты непарметрического теста

median_2021 = medical_cases.query('year==2021')['service_amount'].median()
median_2022 = medical_cases.query('year==2022')['service_amount'].median()

print(f'Медианное значение стоимости обслуживания клиентов в 2021 г. {median_2021} рублей')
print(f'Медианное значение стоимости обслуживания клиентов в 2022 г. {median_2022} рублей')
print(f'Разница между 2022 и 2021 годом {median_2022-median_2021} рублей')
print(f'p-значение {results.pvalue}')
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу, что медианные значения стоимости одинаковые")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

**Вывод**:
* Медианное значение обслуживания клиентов достоверно выросло в 2022 году на 441 рубль.

1.	Средняя стоимость обслуживания пациентов изменилась из-за изменения цен на услуги.
2.	Средняя стоимость обслуживания изменилась из-за изменения среднего количества оказанных услуг на один кейс.
3.	Средняя стоимость обслуживания пациентов изменилась из-за изменений в предпочтении мужчин или женщин, а также из-за различий в предпочтений разных возрастных групп получения медицинского обслуживания.
